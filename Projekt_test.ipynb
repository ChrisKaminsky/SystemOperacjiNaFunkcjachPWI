```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 15
x = np.arange(0, 16,0.1)

# Calculate y values using the function f(x) = 2^x
y = 2**x

# Create the plot
plt.plot(x, y, linestyle='-', label='f(x) = 2^x')
plt.title('Plot of f(x) = 2^x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Highlight a specific range, for example, from x=4 to x=10
a = 4
b = 10
plt.fill_between(x, y, where=(x >= a) & (x <= b), color='red', alpha=0.3, label='Highlighted Area')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
